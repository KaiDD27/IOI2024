```python
import matplotlib.pyplot as plt
import numpy as np

h = [6, 7, 9, 10]
t = [5, 10]

fig, ax = plt.subplots(figsize=(10, 6))

colors = ['red', 'green', 'blue', 'purple']
for i, height in enumerate(reversed(h)):
    throws = np.arange(height, 0, -1)
    ax.scatter(np.cumsum(np.arange(len(throws))), throws, color=colors[i], label=f'Pinecone {4-i}')
    ax.plot(np.cumsum(np.arange(len(throws))), throws, color=colors[i])

ax.set_xlabel('Cumulative throws')
ax.set_ylabel('Height')
ax.set_title('Pinecone Throwing Strategy')
ax.legend()

for throw in t:
    ax.axvline(x=throw, color='black', linestyle='--', label=f'Scenario: {throw} throws')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

h = [1, 3, 4, 6]
t = [5, 10]

fig, ax = plt.subplots(figsize=(10, 6))

colors = ['red', 'green', 'blue', 'purple']
cumulative_throws = 0
for i, height in enumerate(h[::-1]):
    if i == 0:
        throws = np.arange(height, 0, -1)
    else:
        throws = np.arange(height, h[-i-2], -1)
    x = np.arange(cumulative_throws, cumulative_throws + len(throws))
    ax.scatter(x, throws, color=colors[i], label=f'Pinecone {len(h)-i}')
    ax.plot(x, throws, color=colors[i])
    cumulative_throws += len(throws)

ax.set_xlabel('Cumulative throws')
ax.set_ylabel('Height')
ax.set_title('Pinecone Throwing Strategy')
ax.legend()

for throw in t:
    ax.axvline(x=throw, color='black', linestyle='--', label=f'Scenario: {throw} throws')

plt.tight_layout()
plt.show()
```
